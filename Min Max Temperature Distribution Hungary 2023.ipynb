```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
from joypy import joyplot
from pandas.api.types import CategoricalDtype

warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel("temperatures Hungary 2023.xlsx")

df['Month'] = df['Month'].astype(CategoricalDtype(df['Month'].unique()))
```

```python
df
```

```python
plt.figure()

ax, fig = joyplot(
    data=df[['min_temp', 'max_temp', 'Month']], 
    by='Month',
    column=['min_temp', 'max_temp'],
    color=['#2471A3', '#fe7c73'],
    alpha=0.75,
    figsize=(12, 12),
)
ax.suptitle('Min Max Temperature Distribution Hungary 2023', fontsize=20)

plt.show()
```

```python
ax.savefig('Min Max Temperature Distribution Hungary 2023.pdf', format='pdf', dpi=300)
```
